```python
# üß© B√†i t·∫≠p 4.4 - X√°c ƒë·ªãnh t·ª´ ƒëi·ªÉn v√† c√°c vector ƒë·∫∑c tr∆∞ng b·∫±ng ph∆∞∆°ng ph√°p Bag-of-Words

# 1Ô∏è‚É£ Import th∆∞ vi·ªán
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 2Ô∏è‚É£ ƒêo·∫°n vƒÉn
sentences = [
    "Trong m·ªôt khu v∆∞·ªùn r·ªông, c√≥ nhi·ªÅu ng∆∞·ªùi th√≠ch nh·ªØng ho·∫°t ƒë·ªông kh√°c nhau.",
    "Lan y√™u th√≠ch vi·ªác v·∫Ω tranh v√† ƒë·ªçc s√°ch.",
    "An c≈©ng th√≠ch v·∫Ω tranh, nh∆∞ng b·∫°n ·∫•y kh√¥ng th√≠ch ƒë·ªçc s√°ch.",
    "Hoa kh√¥ng th√≠ch v·∫Ω tranh, nh∆∞ng b·∫°n ·∫•y r·∫•t th√≠ch ƒë·ªçc s√°ch v√† ch∆°i piano.",
    "Ho√†ng kh√¥ng h·ªÅ th√≠ch c·∫£ hai ho·∫°t ƒë·ªông n√†y, thay v√†o ƒë√≥ b·∫°n ·∫•y th√≠ch ch∆°i th·ªÉ thao v√† ƒëi du l·ªãch."
]

# 3Ô∏è‚É£ Kh·ªüi t·∫°o m√¥ h√¨nh Bag-of-Words
vectorizer = CountVectorizer()

# 4Ô∏è‚É£ Hu·∫•n luy·ªán v√† bi·∫øn ƒë·ªïi d·ªØ li·ªáu
X = vectorizer.fit_transform(sentences)

# 5Ô∏è‚É£ L·∫•y danh s√°ch t·ª´ trong t·ª´ ƒëi·ªÉn
vocab = vectorizer.get_feature_names_out()

# 6Ô∏è‚É£ Chuy·ªÉn th√†nh DataFrame ƒë·ªÉ hi·ªÉn th·ªã ƒë·∫πp
df = pd.DataFrame(X.toarray(), columns=vocab)

# 7Ô∏è‚É£ Hi·ªÉn th·ªã k·∫øt qu·∫£
print("===== T·ª™ ƒêI·ªÇN (VOCABULARY) =====")
print(vocab)
print("\n===== MA TR·∫¨N BAG-OF-WORDS =====")
print(df)
```
