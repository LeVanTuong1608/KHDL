```python

# Load d liu
iris = load_iris()

# Xem cc keys c trong iris (ty chn)
print(iris.keys())

# Chuyn sang DataFrame cho d c
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# i tn ct cho d c (tu chn)
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# In 5 dng u tin  kim tra
print(df.head())

# V scatter plot: petal_length vs sepal_length
plt.figure(figsize=(8,6))
species = df['species'].unique()
markers = ['o', 's', 'D']  # circle, square, diamond

for sp, m in zip(species, markers):
    sub = df[df['species'] == sp]
    plt.scatter(sub['sepal_length'], sub['petal_length'],
                label=sp, alpha=0.8, marker=m, s=60)

plt.xlabel('Sepal length (cm)')
plt.ylabel('Petal length (cm)')
plt.title('Iris: Sepal length vs Petal length (colored by species)')
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load d liu ch s (0-5) MNIST  c vector ha (54 chiu)
from sklearn.datasets import load_digits

# Ti d liu digits vi 6 lp (0-5)
digits = load_digits(n_class=6)  # Mi nh l 1 vector 64 chiu (8x8), khng phi 54 chiu
labels = digits.target

print('Shape of features:', features.shape)
print('Shape of labels:', labels.shape)
print('Unique labels:', set(labels))

# Hin th 5 mu u tin
print('First 5 feature vectors:')
print(features[:5])
print('First 5 labels:')
print(labels[:5])
```

```python
# Gim chiu d liu v 2 chiu bng PCA v trc quan ha
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Ly li features t digits (nu cha c)
features = digits.data
labels = digits.target

# Gim v 2 chiu
pca = PCA(n_components=2)
features_2d = pca.fit_transform(features)

plt.figure(figsize=(8,6))
scatter = plt.scatter(features_2d[:,0], features_2d[:,1], c=labels, cmap='tab10', alpha=0.3)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('D liu ch s (0-5) MNIST sau khi gim v 2 chiu bng PCA')
plt.colorbar(scatter, ticks=range(6), label='Digit label')
plt.grid(True)
plt.show()
```
