```python

# Load dữ liệu
iris = load_iris()

# Xem các keys có trong iris (tùy chọn)
print(iris.keys())

# Chuyển sang DataFrame cho dễ đọc
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Đổi tên cột cho dễ đọc (tuỳ chọn)
df.rename(columns={
    'sepal length (cm)': 'sepal_length',
    'sepal width (cm)': 'sepal_width',
    'petal length (cm)': 'petal_length',
    'petal width (cm)': 'petal_width'
}, inplace=True)

# In 5 dòng đầu tiên để kiểm tra
print(df.head())

# Vẽ scatter plot: petal_length vs sepal_length
plt.figure(figsize=(8,6))
species = df['species'].unique()
markers = ['o', 's', 'D']  # circle, square, diamond

for sp, m in zip(species, markers):
    sub = df[df['species'] == sp]
    plt.scatter(sub['sepal_length'], sub['petal_length'],
                label=sp, alpha=0.8, marker=m, s=60)

plt.xlabel('Sepal length (cm)')
plt.ylabel('Petal length (cm)')
plt.title('Iris: Sepal length vs Petal length (colored by species)')
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load dữ liệu chữ số (0-5) MNIST đã được vector hóa (54 chiều)
from sklearn.datasets import load_digits

# Tải dữ liệu digits với 6 lớp (0-5)
digits = load_digits(n_class=6)  # Mỗi ảnh là 1 vector 64 chiều (8x8), không phải 54 chiều
labels = digits.target

print('Shape of features:', features.shape)
print('Shape of labels:', labels.shape)
print('Unique labels:', set(labels))

# Hiển thị 5 mẫu đầu tiên
print('First 5 feature vectors:')
print(features[:5])
print('First 5 labels:')
print(labels[:5])
```

```python
# Giảm chiều dữ liệu về 2 chiều bằng PCA và trực quan hóa
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Lấy lại features từ digits (nếu chưa có)
features = digits.data
labels = digits.target

# Giảm về 2 chiều
pca = PCA(n_components=2)
features_2d = pca.fit_transform(features)

plt.figure(figsize=(8,6))
scatter = plt.scatter(features_2d[:,0], features_2d[:,1], c=labels, cmap='tab10', alpha=0.3)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Dữ liệu chữ số (0-5) MNIST sau khi giảm về 2 chiều bằng PCA')
plt.colorbar(scatter, ticks=range(6), label='Digit label')
plt.grid(True)
plt.show()
```
