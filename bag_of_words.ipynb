```python
# 🧩 Bài tập 4.4 - Xác định từ điển và các vector đặc trưng bằng phương pháp Bag-of-Words

# 1️⃣ Import thư viện
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 2️⃣ Đoạn văn
sentences = [
    "Trong một khu vườn rộng, có nhiều người thích những hoạt động khác nhau.",
    "Lan yêu thích việc vẽ tranh và đọc sách.",
    "An cũng thích vẽ tranh, nhưng bạn ấy không thích đọc sách.",
    "Hoa không thích vẽ tranh, nhưng bạn ấy rất thích đọc sách và chơi piano.",
    "Hoàng không hề thích cả hai hoạt động này, thay vào đó bạn ấy thích chơi thể thao và đi du lịch."
]

# 3️⃣ Khởi tạo mô hình Bag-of-Words
vectorizer = CountVectorizer()

# 4️⃣ Huấn luyện và biến đổi dữ liệu
X = vectorizer.fit_transform(sentences)

# 5️⃣ Lấy danh sách từ trong từ điển
vocab = vectorizer.get_feature_names_out()

# 6️⃣ Chuyển thành DataFrame để hiển thị đẹp
df = pd.DataFrame(X.toarray(), columns=vocab)

# 7️⃣ Hiển thị kết quả
print("===== TỪ ĐIỂN (VOCABULARY) =====")
print(vocab)
print("\n===== MA TRẬN BAG-OF-WORDS =====")
print(df)
```
